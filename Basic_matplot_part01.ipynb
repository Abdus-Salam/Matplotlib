# Basic matplotlib tutorial part 02

```python
from numpy import *
from matplotlib import pyplot as plt
```

```python
t = arange(0.0, 5.0, 0.01)
s = cos(2*pi*t)
plt.plot(t, s, lw=2)

# xy (3, 1) : position of the arrow in x and y
# xytext(4, 1.5) : position in the x and y text

plt.annotate('local max', xy = (3, 1), xytext = (4, 1.5), arrowprops = dict(facecolor = 'black', shrink = 0.05))
plt.ylim(-2, 2)
plt.xlim(0, 5.5)
plt.show()
```

```python
x = linspace(0, 15, 100) # total 100 points between 0 to 15
y = random.random(100)

plt.fill_between(x, y, color = '#e2e2e2')

plt.ylim(-2, 2)
plt.xlim(0, 30)
plt.show()
```

```python
x = linspace(0, 10, 200)
y1 = 2 * x + 1
y2 = 3 * x + 1.2
y_mean = 0.5 * x * cos(2*x) + 2.5 * x + 1.1

fig, ax = plt.subplots()

# Plot the envelope with `fill_between`
ax.fill_between(x, y1, y2, color='yellow')

# Plot the "centerline" with `plot`
ax.plot(x, y_mean, color='black')

plt.show()
```

```python
bars = plt.bar([1, 2, 3, 4], [10, 12, 15, 17])
plt.setp(bars[0], hatch='x', facecolor='w')
plt.setp(bars[1], hatch='xx-', facecolor='orange')
plt.setp(bars[2], hatch='+O.', facecolor='c')
plt.setp(bars[3], hatch='*', facecolor='y')

plt.show()
```

```python
x = [1, 2, 3, 4]
y = [10, 12, 15, 19]

fig, ax = plt.subplots()
ax.plot(x, label='some1')
ax.plot(y, label='some2')
ax.set(ylabel='y', xlabel='x', title='title')
plt.legend(loc = 'upper left') #position of the legend

plt.show()
```

```python
t = arange(0.0, 5.0, 0.01)

y = cos(2*pi*t)
plt.plot(t, y, 'r', marker='o', markerfacecolor='y', mec='g')
plt.margins(0.1)
plt.show()
```

```python
time = arange(0.0, 5.0, .01)
amplitude = sin(2*pi*time)

plt.plot(amplitude, 'y-o')
plt.show()
```

```python
# %load exercises/2.2-vmin_vmax_imshow_and_colorbars.py
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)

# Generate random data with different ranges...
data1 = np.random.random((10, 10))
data2 = 2 * np.random.random((10, 10))
data3 = 3 * np.random.random((10, 10))

# Set up our figure and axes...
fig, axes = plt.subplots(ncols=3, figsize=plt.figaspect(0.5))
fig.tight_layout() # Make the subplots fill up the figure a bit more...
cax = fig.add_axes([0.25, 0.1, 0.55, 0.03]) # Add an axes for the colorbar

# Now you're on your own!
for ax, data in zip(axes, [data1, data2, data3]) :
    im = ax.imshow(data, vmin = 0, vmax = 1, interpolation='nearest')

fig.colorbar(im, cax = cax, orientation='horizontal')
plt.show()
```
