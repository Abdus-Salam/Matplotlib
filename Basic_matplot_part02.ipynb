### multiple axes

```python
x = np.linspace(0, 10, 100)
y1, y2, y3, y4 = np.sin(x), np.sin(x + 1), np.sin(x + 2), x*x

fig, axes = plt.subplots(nrows = 2, ncols = 2)

axes[0, 0].plot(x, y1, 'r-o', color='#ed2323', linewidth=1)
axes[0, 1].plot(x, y2)
axes[1, 0].plot(x, y3)
axes[1, 1].plot(x, y4)


axes[0,0].set(title='sin(x)', )
axes[0,1].set(title='sin(x + 1)')
axes[1,0].set(title='sin(x + 2)')
axes[1,1].set(title='x*x')



plt.show()
```

```python
x = np.linspace(0, 10, 100)

y1, y2, y3, y4 = np.sin(x), np.sin(x+1), np.sin(x+2), x*x

fig, (figure) = plt.subplots(nrows=2, ncols=2)

figure[0, 0].plot(x, y1)
figure[0, 1].plot(x, y2)
figure[1, 0].plot(x, y3)
figure[1, 1].plot(x, y4)

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D, axes3d

fig, ax = plt.subplots(1, 1, subplot_kw={'projection': '3d'})
X, Y, Z = axes3d.get_test_data(0.05)
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)

plt.show()
```
